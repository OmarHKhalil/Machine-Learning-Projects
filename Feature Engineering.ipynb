# Feature Engineering:
> - Feature Preprocessing:
تحديث او تحويل الميزات الموجوده في هذه الحاله لا نقوم بانشاء ميزات جديده ولكن نجعل الميزات اكثر افاده
> > - Feature Transformation: 
تحويل الميزات من غير خطيه الى خطية او تحويل التوزيع الملتوي الى توزيع طبيعي
> > > - Logarithm
> > > - Square / cube
> > > - Square root / cube root
> > > - Reciprocal
> > - Feature Scaling:
> > > - Min Max Scaling : القيم بين 0 و 1
> > > - Standard Scaling : القيم بين 1 و -1
> > - Feature Encoding:
> > > - One Hot Encoding: اذا كانت القيم ذات ترتيب لا تصلح لها
> > > - Label Encoding: مفيدة في القيم ذات ترتيب
> > - Feature Combine Sparse Class: عمود فئه يحتوي على 100 فئة مختلفة
> > > - الفئه التي تكرارها اقل يمكن دمجها تحت فئه جديده

> - Feature Generation:
تولدي ميزات جديده  من الميزات الموجوده

> > > - عن طريق الشجره

# Binning:

```python
stroke_data = pd.read_csv("healthcare-dataset-stroke-data.csv")
stroke_data.head()
```

```python
bins = [0, 12, 19 , 30, 60, 90]
group = ['child', 'teenager', 'young_adult', 'middle_aged', 'senior_citizen']
stroke_data['age_category'] = pd.cut(stroke_data['age'], bins, labels=group)
# هنا نستخدم الدالة pd.cut لتقسيم قيم العمود 'age' إلى الفئات المحددة في bins، ونعطي كل فئة اسم من labels=group.
stroke_data[['age_category', 'age']].head()
```

# Binning used decition tree:

```python
bins = [0, 49.5, 66.5, 77.5, 90]
group = ['Bin_1', 'Bin_2', 'Bin_3', 'Bin_4',]
stroke_data['age_category'] = pd.cut(stroke_data['age'], bins, labels=group)
stroke_data[['age_category', 'age']].head()
```

```python
new_df = pd.DataFrame({
    "year": df['Date'].dt.year,
    "month": df['Date'].dt.month,
    "day": df['Date'].dt.day,
    "hour": df['Date'].dt.hour,
    "dayofyear": df['Date'].dt.dayofyear,
    "week": df['Date'].dt.isocalendar().week,
    "dayofweek": df['Date'].dt.dayofweek,
    "dayofweekname": df['Date'].dt.day_name(),
    "quarter": df['Date'].dt.quarter,
})
new_df.head()
```

```python
complete_data = pd.concat([df, new_df], axis=1)
complete_data.head()
```

```python
complete_data['is_weekday'] = 0
for i in range(0, len(complete_data)):
    if((complete_data['dayofweek'][i] == 5) | (complete_data['dayofweek'][i] == 6)):
        complete_data.loc[i, 'is_weekday'] = 0
    else:
        complete_data.loc[i, 'is_weekday'] = 1
```

هي مجموعه من الجداول

```python
import featuretools as ft
```

```python
df = bigmart
df.shape
```

```python
features = df.drop(columns=['Item_Outlet_Sales'])
y = df['Item_Outlet_Sales']
```

```python
es = ft.EntitySet(id = 'bigmart')
es
```

```python
es.add_dataframe(dataframe_name = 'data_1', dataframe = features,
                make_index = True, index = 'index')
es
```

```python
from featuretools import dfs
feature_matrix, feature_defs = dfs(entityset=es,
                                   target_dataframe_name='data_1',
                                   max_depth=1,
                                   trans_primitives=['add_numeric', 'multiply_numeric'])
feature_matrix.head()
```
