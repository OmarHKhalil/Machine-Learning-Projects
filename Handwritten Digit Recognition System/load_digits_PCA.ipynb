```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df
```

```python
df.describe()
```

```python
x = df
y = dataset.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(x)
```

```python
x_scaled = pd.DataFrame(scaler.transform(x), columns=x.columns)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
x.shape
```

```python
#  استخراج اهم الميزات بهذه النسبه اي كانت الميزات التي سوف تتوصل لها
from sklearn.decomposition import PCA
pca = PCA(0.95) 
x_pca = pca.fit_transform(x)
x_pca.shape
```

```python
#  كل مكون بلتقط النسبه من التباين او المعلومات المفيده من البيانات
pca.explained_variance_ratio_
```

```python
# إذا القيمة قريبة جدًا من 0.95 (أو القيمة التي اخترتها)، فهذا يعني انه حافظ على معظم المعلومات.
print(sum(pca.explained_variance_ratio_))
```

```python
#  عدد المكونات
pca.n_components_
```

```python
pca.components_
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
