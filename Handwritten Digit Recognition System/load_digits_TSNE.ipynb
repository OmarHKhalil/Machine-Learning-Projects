```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
x = digits.data
y = digits.target
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=50, random_state=42)
x_pca = pca.fit_transform(x_scaled)
```

```python
pca.explained_variance_ratio_.sum()
```

```python
from sklearn.manifold import TSNE
tsn = TSNE(n_components=2, perplexity=30, learning_rate=200, n_iter=1000, random_state=42, verbose=1)
```

```python
x_tsne = tsn.fit_transform(x_pca)
x_tsne
# يعتبر جيد 0.820507
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(x= x_tsne[:,0], y=x_tsne[:,1], hue=y, palette='tab20')
plt.show() 
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x_tsne, y)
```

```python
x_tsne2 = tsn.fit_transform(x_scaled)
```

```python
sns.scatterplot(x= x_tsne2[:,0], y=x_tsne2[:,1], hue=y, palette='tab20')
```

```python
silhouette_score(x_tsne2, y)
```
