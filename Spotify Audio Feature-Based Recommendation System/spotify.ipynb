```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_songs = pd.read_csv('genres_v2.csv')
spotify_songs.head()
```

```python
spotify_songs.describe()
```

```python
print(spotify_songs.columns,spotify_songs.dtypes)
```

```python
spotify_songs.isnull().sum()
```

```python
spotify_songs.drop(columns=['uri','id','track_href','analysis_url','Unnamed: 0','title','song_name'],axis = 1,inplace = True)
```

```python
spotify_songs = spotify_songs.dropna()
spotify_songs = spotify_songs.drop_duplicates()
spotify_songs.head(5)
```

```python
len(spotify_songs['type'].unique())
```

```python
len(spotify_songs['genre'].unique())
```

```python
from sklearn.preprocessing import LabelEncoder 
columns_to_label_encode = ['type','genre']
le = LabelEncoder()
for col in columns_to_label_encode:
    spotify_songs[col] = le.fit_transform(spotify_songs[col])
    
spotify_songs.head(3)
```

```python
list(le.classes_)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='danceability',y='genre',data=spotify_songs)
plt.title('Genre vs Danceability')
plt.xlabel('Danceability')
plt.ylabel('Genre')
plt.show()
```

```python
le.transform(le.classes_)
```

```python
spotify_songs.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaled_values = scaler.fit_transform(spotify_songs)
scaled_df = pd.DataFrame(scaled_values, columns = spotify_songs.columns)
scaled_df.head(3)
```

```python
from sklearn.cluster import KMeans
inertia = []
range_ = range(1,10)
for i in range_:
    km = KMeans(n_clusters= i).fit(spotify_songs)
    inertia.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 10), inertia)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')  # Within cluster sum of squares
plt.show()
```

```python
km = KMeans(n_clusters = 4)
```

```python
y_pred = km.fit_predict(spotify_songs)
y_pred
```

```python
spotify_songs['label'] = km.labels_
```

```python
spotify_songs.head(5)
```

```python
sns.scatterplot(x='liveness', y='danceability', data=spotify_songs, hue='label', palette='Set1', s=50) 
plt.title('liveness vs danceability')
plt.xlabel('liveness')
plt.ylabel('danceability')
# Visualizing the clusters
centers = km.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, marker='x')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(spotify_songs, y_pred)
```
