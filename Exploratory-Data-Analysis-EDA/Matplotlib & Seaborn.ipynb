```python
#     ما هو متوسط حجم المحرك لكل نوع هيكل سياره
plt.figure(figsize = (10,5))
sns.barplot(data = df, y = 'carbody', x ='enginesize', estimator = 'median', errorbar = 'sd', palette = 'muted')
#  estimator  هي خاصية تحدد الإحصائيات التي تُستخدم لحساب قيم القياسات المُعروضة في المخطط الشريطي

plt.xlabel('mean engine size')
plt.ylabel('car body type')
plt.title('Avarage engine size by car body type')
plt.show()
```

```python
#   يمكن عدم استخدام هذه الطريقه يكفي الطريقه السابقه و تعتبر هذه الطريقه ليست جيده
#   كيف يختلف حجم المحرك باختلاف هيكل السياره مقسم على عدد الابواب
plt.figure(figsize = (6,4))
bar_plot = sns.barplot(data = df, y = 'carbody', x ='enginesize', hue='doornumber', estimator = 'mean', errorbar = 'sd', palette = 'hls')
bar_plot.set_xlabel('mean engine size')
bar_plot.set_ylabel('car body type')
bar_plot.set_title('Avarage engine size by car body type')
plt.show()
```

# 3-Cat plot

```python
#   افضل في التفسير
plt.figure(figsize = (6,4))
bar_plot = sns.catplot(data = df, kind = 'bar', y = 'carbody', x ='enginesize', col='doornumber', estimator = 'mean', errorbar = 'sd')
#  kind ما هو نوع المخطط الذي تريد رسمه 
#  col  تشيه عمل السابقه لكنها تفصل حسب العامود الى اكثر من رسمه
bar_plot.set_axis_labels('mean engine size','car body type')
bar_plot.fig.suptitle('Avarage engine size by car body type split over number of doors', y=1.52)
#   y    هو المسافه عن العنوان
plt.show()
```

# 4-Stacked bar graph
# التكديس

```python
#  كيف يختلف حجم المحرك عبر انواع مختلفه من هياكل السيارات المقسمه حسب عدد الابواب
pivot_df = df.pivot_table(index = 'carbody', columns = 'doornumber', values = 'enginesize', aggfunc = 'mean', fill_value = 0)
#في هذا النموذج المحوري، تتم محاصرة القيم 'enginesize' بناءً على 'carbody' و 'doornumber'، ويُحتسب المتوسط (aggfunc='mean') لهذه القيم.
#وتُستخدم fill_value=0 لتعيين قيمة افتراضية (في هذه الحالة صفر) لأي خلية تكون فارغة بسبب عدم وجود قيمة.
pivot_df#     تختلف عن السابقه بحيث هذه اكثر تخصيصا ويمكن تطبيق عمليات حسابيه عليها بحيث الاخرى هي قثط عباره عن تقاطع القيم بين عموديين
```

```python
pivot_df.plot(kind = 'bar' , stacked = True)
plt.xlabel('car body type')
plt.ylabel('mean engine size')
plt.title('Avarage engine size by car body type split over number of doors')
plt.xticks(rotation = 45)
plt.show()
```

```python
pivot_df.plot(kind = 'barh' , stacked = True, alpha = 0.5)
plt.xlabel('car body type')
plt.ylabel('mean engine size')
plt.title('Avarage engine size by car body type split over number of doors')
plt.yticks(rotation = 45)
plt.show()
```

# Pair plot

    يشبه التمثيل السابق وهو فقط للارقام
ماذا تفعل؟  
- تعرض لك شبكة من الرسوم البيانية حيث كل مربع يمثل العلاقة بين عمودين:  
  - على القطر الرئيسي (diagonal) ستجد مخططات توزيع (histograms أو KDE) لكل عمود على حدة.  
  - خارج القطر ستجد مخططات استطراد (scatter plots) لكل زوج من الأعمدة، مما يسمح برؤية العلاقات أو الترابطات بين المتغيرات.

ما فائدتها؟  
- تساعد في فهم العلاقات المتبادلة بين المتغيرات بسهولة وبصورة بصرية واضحة.  
- تجعل اكتشاف الارتباطات، التوزيعات، أو المجموعات المختلفة في البيانات أسهل جداً.  
- مفيدة لتحليل البيانات الاستكشافي (EDA) قبل بناء نماذج تعلم الآلة.
