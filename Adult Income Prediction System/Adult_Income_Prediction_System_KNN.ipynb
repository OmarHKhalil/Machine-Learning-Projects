```python
#Outliers processing
#The IsolationForest algorithm gave better results than determining (Lower Bound and Upper Bound).
```

```python
columns_to_plot = ['Age', 'Salary', 'Education_Level', 'Capital_Gain', 'Hours_per_Week']
for column in columns_to_plot:
    plt.figure()
    plt.boxplot(df[column], vert=False)
    plt.ylabel(column)
    plt.xlabel('Values')
    plt.title(f'Box Plot for {column}')
    plt.show()
```

```python
# calculate summary statistics 
d=df.select_dtypes(include=[np.number])
df1 = pd.DataFrame()
for column in columns_to_plot:
    mean = d[column].mean() 
    std = d[column].std() 
    
    # Calculate the lower and upper bounds 
    lower_bound = mean - std * 2
    upper_bound = mean + std * 2
    
    print('Lower Bound for', column, ':', lower_bound) 
    print('Upper Bound for', column, ':', upper_bound) 
    print("-------------")
    # Drop the outliers
    df1[column] = d[(d[column] >= lower_bound) & (d[column] <= upper_bound)][column]
#  nanعند حصر المجال بهذه الطريقه وحذف القيم خارج هذا المجال سيؤدي الي وجود قيم 
#  IsolationForest على عكس 
```

```python
from sklearn.preprocessing import MinMaxScaler 
  
df4 = pd.DataFrame()
scaler = MinMaxScaler()
scaled_array = scaler.fit_transform(df3)
df4 = pd.DataFrame(scaled_array, columns=df3.columns)
df4.head()
```

```python
X = df4.loc[:, df4.columns != 'Income_enc']
Y = df4['Incom_enc']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)
```

```python
#Now we will try three values ​​of K.
#1)K=1
#2)K=3
#3)The Best value of K (WE SELECT IT)
```

```python
#1)Model training K=1
```

```python
from sklearn.metrics import classification_report,\
    confusion_matrix
from sklearn.neighbors import KNeighborsClassifier


knn1 = KNeighborsClassifier(n_neighbors=1)
knn1.fit(X_train, y_train)
pred1 = knn1.predict(X_test)
```

```python
# Predictions and Evaluations
# Let's evaluate our KNN model !
from sklearn.metrics import accuracy_score 

accuracy = accuracy_score(y_test, pred1)

print('Accuracy = ', accuracy)
print(confusion_matrix(y_test, pred1))
print(classification_report(y_test, pred1))
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_test, pred1), display_labels = [0, 1])

cm_display.plot()
plt.show()
```

```python
#2)Model training K = 3
```

```python
knn2 = KNeighborsClassifier(n_neighbors=3)
knn2.fit(X_train, y_train)
pred2 = knn2.predict(X_test)
```

```python
accuracy2 = accuracy_score(y_test, pred2)

print('Accuracy = ', accuracy2)
print(confusion_matrix(y_test, pred2))
print(classification_report(y_test, pred2))
```

```python
#3)Choose the best value for K
```

```python
error_rate = []
# Will take some time
for i in range(1,20):

    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))


plt.plot(range(1,20),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=8)
plt.title('Error Rate vs. K Value', fontsize=20)
plt.xlabel('K',fontsize=15)
plt.ylabel('Error (misclassification) Rate',fontsize=15)
```

```python
print("Minimum error:",min(error_rate),"at K =",error_rate.index(min(error_rate)))
K = error_rate.index(min(error_rate))
```

```python
knn3 = KNeighborsClassifier(n_neighbors= K)
knn3.fit(X_train, y_train)
pred3 = knn3.predict(X_test)
```

```python
accuracy3 = accuracy_score(y_test, pred3)

print('Accuracy = ', accuracy3)
print(confusion_matrix(y_test, pred3))
print(classification_report(y_test, pred3))
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix(y_test, pred3), display_labels = [0, 1])

cm_display.plot()
plt.show()
```
