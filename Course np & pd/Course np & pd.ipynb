# lec 4: Slicing

```python
# a = b[2:3]   b يؤدي الى تغيير في  a اي تغيير في 
# a= b[2:3].copy()   هكذا نحل المشكله
print(numbers)
print(numbers[4:8])
print(numbers[:8])
print(numbers[4:])
print(numbers[::-1])
print(numbers[-1::])
print("jumb: ",numbers[::2]) # يقفز خطوتين 
print("jumb: ",numbers[::-2]) # يقفز خطوتين لكن يعكس المصفوفه
```

```python
#   اذا كان غير موجودfalse اذا وجد العنصر وtrue  يتحقق اذا كان العنصر موجود او لا لكل عنصر ويضع  
print(numbers == 5)
# يرجع الفهرس الذي فيه عنصر محدد
idx = np.argwhere(numbers ==5)[0][0]
print("The index of 5 is: ",int(idx)+1)
```

# lec 6: more indexing

```python
# slicingالميزه الرئيسيه فيها هي عند التعديل على المصفوفه لا يتم التعديل على المصفوفه القديمه على عكس 
a = np.arange(10)
print(a)
# تعمل على اخذ عناصر محدد من المصفوفه ووضعهم في مصفوفه جديده
print(a[[1,5,9]])
# يجب وضعهم لجميع عناصر المصفوفه
print(a[[True,True,False,True,False,True,True,False,True,False]])
print(a[a<8],a[a==8],a[(a<8)&(a>4)])
#    اذا وجد عنصر فقط  and  تستخدم 
# تستخدم & اذا وجد اكثر من عنصر مصفوفه
# | , or , ~ , not مشابه لما سبق
```

```python
arr1 = np.round(np.random.rand(2,3)*10)
print(arr1)
arr2 = np.round(np.random.rand(2,2)*10)
print(arr2)
#(tuple)اذا اردنا دمج الاعمدة مصفوفتين يجب استخدامها بشرط ان يكون عدد الاسطر متشابه ويجب ان يكون الدخل 
arr3 = np.hstack((arr1,arr2))
print(arr3)
```

```python
arr4 = np.round(np.random.rand(2,3)*10)
print(arr4)
arr5 = np.round(np.random.rand(3,3)*10)
print(arr5)
# اذا اردنا دمج الاسطر بشرط ان ايكون عدد الاعمده متساوي
arr6 = np.vstack((arr4,arr5))
print(arr6)
```

```python
marks_dict = {'A+':4,'A-':3.7,'A':3.5,'B+':3.25,'B':3,'B-':2.7}
print("dict1: ",marks_dict)
marks = pd.Series(marks_dict)
print("series: \n",marks)
marks_dict2 = {'A+':100,'A-':96,'A':94,'B+':90,'B':85,'B-':80,'f':0}
print("dict2: ",marks_dict2)
marks2 = pd.Series(marks_dict2)
print("series: \n",marks2)
print(marks['A+':'B'])
print(marks[0:5])
```
